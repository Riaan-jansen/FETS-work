```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
import matplotlib.colors as mcolors

df = pd.read_csv('FETSMETS_csv.csv')

current = df['Current (avg) (mA)']
y = df['N yield (n/s)']
names = df['Facility']
mat = df['Material']
perg = df['Proton beam (MeV)']

mat_colours = []
for i in mat:
    if i == 'Be9':
        mat_colours.append('orange')
    elif i == 'Li7':
        mat_colours.append('cyan')
    else:
        mat_colours.append('grey')


fig, ax = plt.subplots(2, 1, figsize=(8,10))

ax[0].scatter(current, y, s=perg**2+8, c=mat_colours, marker='o') # size of marker and colour indicates proton energy
xbar = [0, 30]
ybar = [1000000000, 30000000000000]
ax[0].plot(xbar, ybar, ls='--')
ax[0].set_yscale('log')


for i, char in enumerate(names):
    ax[0].annotate(char, (current[i], y[i]))#, arrowprops=dict(headwidth=3, width=1))


plt.title('Neutron yield-current by facility.')
plt.xlabel('Current (avg) (mA)')
plt.ylabel('N yield (n/s)')

fig.show()
```
